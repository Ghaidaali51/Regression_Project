```python
car_name = []
car_price = []
car_trans = []
car_petrol = []
car_driven = []
car_feature = []
 
for res in results:
    
    
    try:
        car_name.append(res.find('h2' , {'class' : 'fcN7dZ z2jAym p2jAym b2jAym' }).get_text())
    except:
        car_name.append('n/a')
        
        
        
    try:
        car_price.append(res.find('span' , {'class' : 'd3uM7V C2jAym p2jAym b2jAym' }).get_text())
    except:
        car_price.append('n/a')
        
        
        
        
    try:
        car_trans.append(res.find('span' , {'title' : 'Automatic' }).get_text())
    except:
        car_trans.append('Manual')
        
        
        
        
    try:
        car_petrol.append(res.find('li' , {'class' : 'c3n44x d3n44x' }).get_text())
    except:
        car_feature.append('n/a')
        
        
    try:
        car_driven.append(res.find('li' , {'class' : 'c3n44x d3n44x f3n44x' }).get_text())
    except:
        car_feature.append('n/a')
        
    try:
        car_feature.append(res.find('div' , {'class' : 'b15_Lb c15_Lb' }).get_text())
    except:
        car_feature.append('n/a')
    
        
```
